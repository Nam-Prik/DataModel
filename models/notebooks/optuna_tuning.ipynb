# Optuna Hyperparameter Tuning
Uses Bayesian optimization to find better hyperparameters than GridSearch faster.

```python
import sys
import os
sys.path.append(os.path.abspath('..'))

import pandas as pd
import optuna
from src.data import split_data
from src.tuner import optuna_tune_xgboost, optuna_tune_rf
from src.evaluate import full_evaluation_report
from src.utils import save_model
from src.config import TRAIN_CONFIG
```

```python
df = pd.read_csv('../data/sample_data.csv')
X_train, X_test, y_train, y_test = split_data(df, target_col='target', test_size=TRAIN_CONFIG['test_size'], random_state=TRAIN_CONFIG['random_state'])
feature_names = [c for c in df.columns if c != 'target']
```

### Tune XGBoost with Optuna

```python
xgb_model, xgb_best_params, xgb_study = optuna_tune_xgboost(X_train, y_train)
print('Best XGBoost Params:', xgb_best_params)
```

```python
full_evaluation_report(xgb_model, X_test, y_test, feature_names, 'XGBoost (Optuna)')
```

### Tune Random Forest with Optuna

```python
rf_model, rf_best_params, rf_study = optuna_tune_rf(X_train, y_train)
print('Best RandomForest Params:', rf_best_params)
```

```python
full_evaluation_report(rf_model, X_test, y_test, feature_names, 'RandomForest (Optuna)')
```

### Optuna Optimization History

```python
optuna.visualization.matplotlib.plot_optimization_history(xgb_study)
import matplotlib.pyplot as plt
plt.title('XGBoost Optuna Optimization History')
plt.show()
```
